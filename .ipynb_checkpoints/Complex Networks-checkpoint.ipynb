# 1.1 What are Complex Networks?

A complex network is a mathematical abstraction used to represent systems composed of a large number of interacting units whose collective behavior cannot be trivially reduced to the sum of their parts. Formally, a network is expressed as a graph 

$$G=(V,E),$$


where $V$ is the set of nodes (also called vertices), representing the fundamental elements of the system. Depending on the context, these may correspond to people in a social network, proteins in a biological pathway, computers in the internet, neurons in the brain, cities in an infrastructure system, or even abstract entities such as concepts in a knowledge graph. The set $E$ contains the edges (or links), which represent the relationships or interactions between nodes—friendships, biochemical reactions, communication channels, power lines, trade exchanges, or logical connections.

What makes such a graph complex is not merely the number of its components but the fact that its structure is neither perfectly regular, as in a crystal lattice, nor completely random, as in an Erdős–Rényi model. Instead, empirical networks tend to exhibit non-trivial structural properties that bridge order and randomness. These include:

- the presence of highly connected hubs that dominate connectivity,
- the organization of nodes into tightly knit communities,
- the small-world effect, whereby any two nodes can often be connected through remarkably short paths,
- and heterogeneous degree distributions reflecting diverse interaction patterns.

Such features are not just mathematical curiosities—they give rise to behaviors and emergent phenomena that cannot be predicted by studying individual components in isolation. Epidemics spread faster on networks with hubs, synchronization emerges in neuronal and technological systems, and robustness or fragility of infrastructures often hinges on the topology of connections.

The study of complex networks, therefore, offers a universal framework to analyze a wide variety of systems, from the microscopic scale of molecules and cells to the macroscopic scale of societies, economies, and ecosystems. Despite the diversity of domains, common patterns arise: scale-free structures, modularity, assortativity, and hierarchical organization. These recurring motifs suggest that vastly different systems may be governed by shared organizing principles.

In recent decades, network science has evolved into an interdisciplinary field at the intersection of mathematics, physics, computer science, and data analysis. It provides not only descriptive tools for mapping real-world systems but also predictive and prescriptive frameworks. By quantifying structural metrics—such as degree distribution, clustering, centrality, or path length—researchers can uncover hidden organization, anticipate system-level responses, and design interventions, whether to immunize populations against contagion, prevent cascading failures in power grids, or identify critical nodes in transportation networks.

Ultimately, the study of complex networks reveals that connectivity is the substrate of complexity. By analyzing how nodes interact and how collective patterns arise from these interactions, network science allows us to understand, model, and in some cases control the dynamics of the most intricate systems found in nature and society!

# 1.2 Why Complex Networks Matter?

The study of complex networks provides a universal language for describing systems across disciplines. Instead of treating each system as an isolated entity with its own bespoke rules, we abstract them into nodes and edges and then apply a shared mathematical and computational framework. This approach allows us to detect commonalities in systems that, on the surface, may appear unrelated, and to transfer insights from one field of study to another. In this way, networks serve as a bridge across scientific domains, unifying the study of complexity under a coherent paradigm.

The power of this abstraction lies in its generality. Whether the nodes represent people, proteins, institutions, or galaxies, the same set of structural properties and analytical tools can be employed. Once a system is cast into the language of networks, discoveries in one area can illuminate phenomena in another, revealing underlying organizational principles that cut across nature, technology, and society.

For example:

- **Epidemics and contagion dynamics**. The mathematics developed to describe how diseases spread in populations—taking into account superspreaders, clustering, and mobility—also applies to how computer viruses propagate across the Internet or how rumors and misinformation diffuse through online platforms. Network topology determines thresholds for outbreaks, strategies for immunization, and the speed at which contagion travels.
- **Resilience and fragility**. Studies of power grids, where nodes represent substations and edges correspond to transmission lines, show that the removal of a few highly connected hubs can lead to catastrophic blackouts. This insight directly parallels ecological networks, where the extinction of a keystone species can destabilize entire food webs. In both cases, network analysis helps us anticipate which elements are critical for stability and which are redundant.
- **Collaboration and knowledge diffusion**. The structure of scientific collaboration networks not only sheds light on how researchers form communities but also on how new ideas emerge and spread. Similar models can be applied to study innovation in technology, the diffusion of cultural trends, or even the global distribution of patents and intellectual property. Understanding network connectivity helps explain why some ideas spread globally while others remain confined to local clusters.
- **Transportation and communication**. Networks originally designed to model air traffic, shipping routes, or urban mobility offer deep analogies with neuronal pathways in the brain. Both domains benefit from efficient routing, redundancy to prevent failure, and modular organization to balance local processing with global connectivity. The small-world property, for example, is shared by airline systems and neural circuits, ensuring rapid communication across large distances.
- **Financial and systemic risk**. Financial institutions are connected by lending, borrowing, and trading relationships that form intricate interbank networks. Insights from network analysis have been instrumental in identifying systemic risks, such as cascading defaults that can propagate across the entire financial system. These models are not limited to economics: the same mathematics applies to cascades of failure in technological infrastructures, such as the Internet or water distribution systems.

Beyond these examples, the universality of networks extends to linguistics, sociology, neuroscience, climate science, and even astrophysics. Language networks reveal the structure of grammar and meaning; social networks uncover patterns of influence and inequality; climate networks highlight teleconnections such as El Niño; and cosmic networks describe how galaxies cluster into filaments on the largest observable scales.

Thus, complex networks are not merely descriptive tools. They allow us to:

- Recognize recurring structural patterns such as hubs, communities, and modularity.
- Identify universal mechanisms of growth, adaptation, and resilience.
- Transfer strategies from one domain to another, such as using epidemiological models to mitigate the spread of misinformation.
- Design interventions and predictions by targeting critical nodes or simulating system-wide responses.

In short, complex networks matter because they reveal that connectivity is a fundamental driver of complexity. By framing diverse systems in the same language, network science enables us to move from fragmented, domain-specific insights toward a more integrated understanding of how complex systems function, adapt, and evolve.

# 1.3 Structural Properties in Complex Networks  

Real-world networks are not random; they exhibit recurring structural patterns that emerge consistently across domains as diverse as biology, sociology, infrastructure, and technology. The task of network science is to identify these patterns, describe them systematically, and interpret what they reveal about the functioning and dynamics of the underlying system. To accomplish this, researchers rely on the study of topology—the arrangement of connections among nodes—and on a set of analytical tools that quantify structure in precise ways.

The topology of a network tells us far more than who is connected to whom. It encodes information about efficiency, robustness, hierarchy, and the potential for emergent phenomena such as synchronization, contagion, or cooperation. For instance, a network dominated by a few highly connected hubs behaves very differently from one in which connections are more evenly distributed. Similarly, a network composed of tightly knit clusters will exhibit different diffusion dynamics compared to a network that is largely homogeneous.

To study these topological properties, researchers have developed a toolbox of metrics and approaches, each designed to highlight a specific feature of network structure:

- **Degree and connectivity patterns**. By looking at how many connections each node has, we can detect the presence of hubs, distinguish between homogeneous and heterogeneous networks, and gain insight into potential vulnerabilities or advantages in spreading processes.
- **Clustering and local cohesion**. Clustering measures how tightly groups of nodes are interconnected, capturing the idea of communities or cliques. High clustering often signals the presence of local redundancy, social circles, or functional modules.
- **Path lengths and efficiency**. The concept of distance in a network—how many steps are needed to travel from one node to another—reflects the efficiency of communication or transport within the system. Networks with short path lengths exhibit the so-called small-world property, enabling rapid global reach despite local limitations.
- **Community detection**. Many real networks are not uniform but are divided into clusters or modules that correspond to meaningful subsystems: scientific disciplines, functional regions of the brain, industrial sectors, or cultural groups. Identifying these communities is central to understanding the organization of complexity.
- **Centrality measures**. Not all nodes are equally important. Some act as bridges, others as hubs, and still others as influencers within their communities. Centrality metrics allow us to rank nodes by their structural significance, whether in terms of influence, control of information flow, or vulnerability to failure.
- **Assortativity and correlations**. Networks can be assortative, where similar nodes tend to connect (e.g., people with similar degrees of influence), or disassortative, where opposites attract (e.g., hubs connecting to peripheral nodes). These patterns influence stability, robustness, and dynamics.
- **Hierarchical and multilayer structures**. Many systems are not flat but layered, with interdependent sub-networks operating at different scales—such as the relationship between transportation, communication, and power infrastructures. Recognizing and analyzing these hierarchies is crucial to modeling their resilience.

Beyond these classical metrics, network science also employs visualization techniques and computational algorithms to explore structure. Visualization allows researchers to see patterns—clusters, bridges, bottlenecks—that may not be obvious from raw data. Algorithms for community detection, motif finding, or network growth modeling provide more systematic ways of uncovering hidden organization.

Importantly, these tools are not only descriptive but predictive. By quantifying topological features, researchers can forecast how a network might respond to perturbations, such as the removal of a node, the introduction of new links, or the spread of a contagion. For example, analyzing degree distribution can reveal whether a system is robust against random failures but fragile against targeted attacks, while clustering coefficients can predict the likelihood of local information reinforcement in social systems.

In sum, the study of topology and the application of network metrics give us a lens to translate complexity into measurable form. Through this lens, we can detect regularities that cut across disciplines, compare systems that are otherwise incommensurable, and begin to understand the hidden order underlying seemingly chaotic interconnections. We will go deeper into these structural properties in the following sections.

# 2. Network Metrics

One of the key strengths of **network science** lies in its ability to describe complex systems using **quantitative measures** that go beyond simple visualization.

When we look at a graph, we may recognize hubs, clusters, or short paths, but these impressions are subjective. To study networks rigorously, we need to formalize these patterns with **metrics** that capture structural properties.  

Network metrics provide:  
- **Comparability**: They allow us to compare networks from very different domains (e.g., biological, technological, or social) using the same mathematical language.  
- **Insight into dynamics**: Many dynamical processes—such as epidemics, synchronization, or cascading failures—are deeply shaped by structural properties.  
- **Predictive power**: By quantifying how a network is organized, metrics help anticipate robustness, vulnerability, and spreading capacity.  

Some of the most fundamental metrics include **degree, clustering, path length, modularity, assortativity, and centralities**. Each one emphasizes a different aspect of network organization. Importantly, they are not isolated: together they provide a holistic picture of how local connections give rise to global structure and behavior.  

## 2.1 Degree  

The **degree** of a node is the most basic but also one of the most important descriptors in network analysis. It represents the **number of edges** connected to a node, that is, how many direct neighbors it has.  

Mathematically, for a node \(i\) in a network with adjacency matrix $A$:  

$$k_i = \sum_{j \in V} A_{ij},
$$  

where $A_{ij} = 1$ if node $i$ is connected to node $j$, and $0$ otherwise.  

The degree is important because it give us an idea about:

- **Local measure of importance**: Nodes with high degree—known as *hubs*—often play a disproportionate role in connectivity and information flow.  
- **Foundation for advanced metrics**: Many higher-level measures, such as centrality, assortativity, and clustering, build directly on degree.  
- **Determinant of dynamics**:  
  - In epidemic spreading, highly connected nodes accelerate transmission.  
  - In robustness analysis, networks with hubs can withstand random failures but are fragile to targeted attacks.  
  - In transport or communication systems, hubs act as key relay points, but can also be bottlenecks.  
- **Bridge between local and global structure**: Degree provides the first step to linking local connectivity with global network organization.  

Degree is therefore the **building block metric**: it is simple to define, but it reveals critical insights about how networks function.  

## 3.1 Degree Distribution  

The **degree** of a node $i$ is the number of edges incident upon it:  


$$k_i = \sum_{j \in V} A_{ij},$$
 

where $A_{ij}$ is the adjacency matrix $(A_{ij} = 1$ if $i$ is connected to $j$, and $0$ otherwise).  

The **degree distribution** is the probability \(P(k)\) that a randomly chosen node has degree \(k\). In scale-free networks,  

$$P(k) \sim k^{-\gamma}, \quad \gamma > 1.$$ 

Some common degree distributions observed in real-world networks, with examples, are:

### 3.1.1 Poisson Distribution (Random Networks)  

In **Erdős–Rényi random graphs**, degrees follow a Poisson distribution:  

$$P(k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad \lambda = \langle k \rangle,$$ 

- Most nodes have degree close to the mean.  
- Hubs are extremely unlikely.  
- Example: Some engineered networks, where connections are uniformly distributed.


```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Parameters
n = 1000    # number of nodes
p = 0.01    # edge probability
seed = 42   # reproducibility

# Generate Erdős–Rényi random graph
G = nx.erdos_renyi_graph(n=n, p=p, seed=seed)

# Degree sequence
degrees = [d for _, d in G.degree()]
unique, counts = np.unique(degrees, return_counts=True)

# Normalize to get probabilities
probabilities = counts / counts.sum()

# Plot empirical degree distribution
plt.figure(figsize=(5,4))
plt.bar(unique, probabilities, width=0.9, alpha=0.7, color="steelblue")
plt.title(f"Empirical Degree Distribution (Erdős–Rényi G(n={n}, p={p}))", fontsize=14)
plt.xlabel("Degree k", fontsize=12)
plt.ylabel("P(k)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Print stats
print(f"Expected average degree λ = {p*(n-1):.2f}")
print(f"Empirical average degree = {np.mean(degrees):.2f}")
```

### 3.1.2 Binomial Distribution (Finite Random Networks)  

For a network with $N$ nodes and independent probability $p$ of an edge between any two nodes,  

$$P(k) = \binom{N-1}{k} p^k (1-p)^{N-1-k}.$$ 

- Approximates a Poisson distribution for large $N$ and small $p$.  
- Useful for finite-size Erdős–Rényi graphs.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters (finite N) ---
N = 1_000    # number of nodes
p = 0.01     # independent edge probability
seed = 3405    # reproducibility

# --- Build finite Erdős–Rényi random graph G(N, p) ---
G = nx.erdos_renyi_graph(n=N, p=p, seed=seed, directed=False)

# --- Degree sequence & empirical PMF ---
degrees = np.fromiter((d for _, d in G.degree()), dtype=int)
k_vals, counts = np.unique(degrees, return_counts=True)
empirical_p = counts / counts.sum()

# --- Plot empirical degree distribution (Binomial in finite ER) ---
plt.figure(figsize=(5,4))
plt.bar(k_vals, empirical_p, width=0.9, alpha=0.75)
plt.title(f"Empirical Degree Distribution (Finite ER: G(N={N}, p={p}))")
plt.xlabel("Degree k")
plt.ylabel("P(k)")
plt.tight_layout()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# --- Binomial expectations for sanity checks ---
binom_mean = (N - 1) * p
binom_var  = (N - 1) * p * (1 - p)

print(f"Binomial E[k] = (N-1)p = {binom_mean:.3f}")
print(f"Empirical mean degree = {degrees.mean():.3f}")
print(f"Binomial Var[k] = (N-1)p(1-p) = {binom_var:.3f}")
print(f"Empirical variance = {degrees.var(ddof=0):.3f}")
```

### 3.1.3 Power-Law Distribution (Scale-Free Networks)  

Many real-world networks exhibit a **scale-free degree distribution**, characterized by a heavy tail:  

$$P(k) \sim k^{-\gamma}, \quad \gamma > 1.$$ 

- A few hubs have very high degree, while most nodes have few connections.  
- Example: The Internet, citation networks, airline networks.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters for Barabási–Albert model ---
N = 2000   # number of nodes
m = 3      # number of edges each new node attaches with
seed = 42  # reproducibility

# --- Generate Barabási–Albert scale-free network ---
G = nx.barabasi_albert_graph(n=N, m=m, seed=seed)

# --- Degree sequence & empirical distribution ---
degrees = np.fromiter((d for _, d in G.degree()), dtype=int)
k_vals, counts = np.unique(degrees, return_counts=True)
empirical_p = counts / counts.sum()

# --- Plot empirical degree distribution ---
plt.figure(figsize=(5,4))
plt.scatter(k_vals, empirical_p, alpha=0.7, s=30, color="crimson")
plt.xscale("log")
plt.yscale("log")
plt.title(f"Empirical Degree Distribution (Barabási–Albert, N={N}, m={m})")
plt.xlabel("Degree k (log scale)")
plt.ylabel("P(k) (log scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

print(f"Number of nodes: {N}, edges per new node: {m}")
print(f"Empirical average degree = {degrees.mean():.2f}")
```

### 3.1.4 Exponential Distribution  

In some networks, the probability of high-degree nodes decays exponentially:  

$$P(k) \sim e^{-\alpha k}.
$$  

- Less heterogeneous than power-law.  
- Still allows variability, but hubs are unlikely.  
- Example: Certain social or technological networks with resource constraints. 

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters ---
N = 2000         # number of nodes
scale = 5        # scale parameter for exponential distribution (mean degree)
seed = 42
rng = np.random.default_rng(seed)

# --- Generate degree sequence from exponential distribution ---
degrees = rng.exponential(scale=scale, size=N).astype(int)

# Ensure all degrees >= 1 (configuration_model requires positive even sum)
degrees[degrees < 1] = 1
if degrees.sum() % 2 != 0:  # must be even for configuration model
    degrees[0] += 1

# --- Build random network with exponential degree sequence ---
G = nx.configuration_model(degrees, seed=seed)
G = nx.Graph(G)   # remove parallel edges
G.remove_edges_from(nx.selfloop_edges(G))  # remove self-loops

# --- Empirical degree distribution ---
deg_sequence = [d for _, d in G.degree()]
k_vals, counts = np.unique(deg_sequence, return_counts=True)
empirical_p = counts / counts.sum()

plt.figure(figsize=(5,4))
plt.scatter(k_vals, empirical_p, color="darkgreen", alpha=0.7, s=25)
plt.yscale("log")
plt.title(f"Empirical Degree Distribution (Exponential, N={N})")
plt.xlabel("Degree k")
plt.ylabel("P(k) (log scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

### 3.1.5 Stretched Exponential (Weibull)  

Intermediate between power-law and exponential:  

$$P(k) \sim e^{-\alpha k^{\beta}}, \quad 0 < \beta < 1.
$$  

- Captures networks that are neither purely scale-free nor exponential.  
- Example: Some collaboration and communication networks.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters ---
N = 2000         # number of nodes
scale = 5.0      # scale parameter (λ) for Weibull
beta = 0.6       # shape parameter (0 < β < 1 => stretched exponential)
seed = 42
rng = np.random.default_rng(seed)

# --- Generate degree sequence from Weibull distribution ---
# numpy gives samples from Weibull with PDF ~ (β * x^(β-1)) * exp(-x^β)
# scale it to get desired degree range
degrees = (rng.weibull(beta, size=N) * scale).astype(int)

# Ensure valid sequence for configuration model
degrees[degrees < 1] = 1
if degrees.sum() % 2 != 0:
    degrees[0] += 1

# --- Build random network from degree sequence ---
G = nx.configuration_model(degrees, seed=seed)
G = nx.Graph(G)  # remove parallel edges
G.remove_edges_from(nx.selfloop_edges(G))

# --- Empirical degree distribution ---
deg_sequence = [d for _, d in G.degree()]
k_vals, counts = np.unique(deg_sequence, return_counts=True)
empirical_p = counts / counts.sum()

plt.figure(figsize=(5,4))
plt.scatter(k_vals, empirical_p, color="purple", alpha=0.7, s=25)
plt.yscale("log")
plt.title(f"Empirical Degree Distribution (Weibull, β={beta}, N={N})")
plt.xlabel("Degree k")
plt.ylabel("P(k) (log scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

print(f"Target Weibull β = {beta}, scale = {scale}")
print(f"Empirical mean degree = {np.mean(deg_sequence):.2f}")
```

### 3.1.6 Log-Normal Distribution  

Another common heavy-tailed distribution:  

$$P(k) = \frac{1}{k \sigma \sqrt{2\pi}} \exp \left( - \frac{(\ln k - \mu)^2}{2\sigma^2} \right),
$$ 

where $\mu$ and $\sigma$ are the mean and standard deviation of $\ln k$.  

- Degree distribution arises from **multiplicative growth processes**.  
- Example: Wealth distributions, metabolic networks, and some online networks.  



```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters ---
N = 2000         # number of nodes
mu = 2.0         # mean of log(degree)
sigma = 0.8      # std dev of log(degree)
seed = 42
rng = np.random.default_rng(seed)

# --- Generate degree sequence from log-normal distribution ---
degrees = rng.lognormal(mean=mu, sigma=sigma, size=N).astype(int)

# Ensure validity for configuration model
degrees[degrees < 1] = 1
if degrees.sum() % 2 != 0:
    degrees[0] += 1

# --- Build random network from degree sequence ---
G = nx.configuration_model(degrees, seed=seed)
G = nx.Graph(G)  # remove parallel edges
G.remove_edges_from(nx.selfloop_edges(G))

# --- Empirical degree distribution ---
deg_sequence = [d for _, d in G.degree()]
k_vals, counts = np.unique(deg_sequence, return_counts=True)
empirical_p = counts / counts.sum()

# --- Plot empirical distribution ---
plt.figure(figsize=(5,4))
plt.scatter(k_vals, empirical_p, color="orange", alpha=0.7, s=25)
plt.xscale("log")
plt.yscale("log")
plt.title(f"Empirical Degree Distribution (Log-Normal, μ={mu}, σ={sigma}, N={N})")
plt.xlabel("Degree k (log scale)")
plt.ylabel("P(k) (log scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

print(f"Target log-normal parameters: μ={mu}, σ={sigma}")
print(f"Empirical mean degree = {np.mean(deg_sequence):.2f}")
print(f"Empirical max degree = {np.max(deg_sequence)}")
```

## 3.2 Clustering Coefficient  

The **local clustering coefficient** of node $i$ measures how close its neighbors are to forming a complete clique:  

$$C_i = \frac{2 e_i}{k_i (k_i - 1)},
$$ 

where $e_i$ is the number of edges among the neighbors of $i$.  

The **average clustering coefficient** of the network is:  

$$C = \frac{1}{N} \sum_{i=1}^N C_i.
$$  

**Example:** In social networks, “friends of friends” are often themselves friends.  


```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters ---
N = 300
k_target = 8
p_er = k_target / (N - 1)   # ER edge probability for ~k_target average degree
k_ws = k_target             # WS ring neighbors (must be even and < N)
beta_ws = 0.05              # small rewiring -> high clustering
seed = 42

# --- Build graphs ---
G_lowC  = nx.erdos_renyi_graph(n=N, p=p_er, seed=seed)                 # low clustering
G_highC = nx.watts_strogatz_graph(n=N, k=k_ws, p=beta_ws, seed=seed)   # high clustering

# --- Compute average clustering for titles ---
Cavg_low  = nx.average_clustering(G_lowC)
Cavg_high = nx.average_clustering(G_highC)

# --- Layouts (fixed seeds for reproducibility) ---
pos_low  = nx.spring_layout(G_lowC, seed=seed)        # force-directed layout
pos_high = nx.spring_layout(G_highC, seed=seed + 1)   # separate layout seed

# --- Plot graphs side-by-side ---
plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
nx.draw(
    G_lowC, pos_low,
    node_size=20, width=0.3, alpha=0.8, with_labels=False
)
ax1.set_title(f"Low Clustering (Erdős–Rényi)\nAverage C = {Cavg_low:.3f}")
ax1.axis("off")

ax2 = plt.subplot(1, 2, 2)
nx.draw(
    G_highC, pos_high,
    node_size=20, width=0.3, alpha=0.8, with_labels=False
)
ax2.set_title(f"High Clustering (Watts–Strogatz, β={beta_ws})\nAverage C = {Cavg_high:.3f}")
ax2.axis("off")

plt.tight_layout()
plt.show()

deg_low  = np.mean([d for _, d in G_lowC.degree()])
deg_high = np.mean([d for _, d in G_highC.degree()])
print(f"ER    avg degree ≈ {deg_low:.2f},   avg clustering = {Cavg_low:.3f}")
print(f"WS    avg degree ≈ {deg_high:.2f},  avg clustering = {Cavg_high:.3f}")
```

## 3.3 Small-World Property (Average Path Length)  

The **shortest path length** between nodes $i$ and $j$ is denoted $d(i,j)$. The **average path length** of the network is:  

$$L = \frac{1}{N(N-1)} \sum_{i \neq j} d(i,j).
$$ 

**Example:** Despite its size, the World Wide Web exhibits paths of only a few clicks between websites.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Parameters ---
N = 300          # number of nodes
k = 8            # each node connected to k nearest neighbors on the ring (k must be even)
beta = 0.1       # rewiring probability for WS (0 -> lattice; small -> small-world)
seed = 42

# --- Build graphs ---
# Regular ring lattice (Watts–Strogatz with beta=0)
G_lattice = nx.watts_strogatz_graph(n=N, k=k, p=0.0, seed=seed)

# Small-world: rewire a small fraction of edges
G_smallworld = nx.watts_strogatz_graph(n=N, k=k, p=beta, seed=seed)

# --- Metrics: average path length (L) and average clustering (C) ---
# These graphs are connected for typical (N, k), so L is well-defined
L_lattice = nx.average_shortest_path_length(G_lattice)
C_lattice = nx.average_clustering(G_lattice)

L_small   = nx.average_shortest_path_length(G_smallworld)
C_small   = nx.average_clustering(G_smallworld)

# --- Layouts ---
pos_lattice = nx.circular_layout(G_lattice)             # show the ring structure
pos_small   = nx.spring_layout(G_smallworld, seed=seed) # show shortcut-induced compression

# --- Plot: draw graphs themselves, side-by-side ---
plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
nx.draw(G_lattice, pos_lattice, node_size=18, width=0.4, alpha=0.9, with_labels=False)
ax1.set_title(f"Regular Ring Lattice (β=0)\nL = {L_lattice:.2f},  C = {C_lattice:.2f}")
ax1.axis("off")

ax2 = plt.subplot(1, 2, 2)
nx.draw(G_smallworld, pos_small, node_size=18, width=0.4, alpha=0.9, with_labels=False)
ax2.set_title(f"Small-World (Watts–Strogatz, β={beta})\nL = {L_small:.2f},  C = {C_small:.2f}")
ax2.axis("off")

plt.tight_layout()
plt.show()

avg_deg_lattice = np.mean([d for _, d in G_lattice.degree()])
avg_deg_small   = np.mean([d for _, d in G_smallworld.degree()])
print(f"Lattice:      avg degree ≈ {avg_deg_lattice:.2f}, L = {L_lattice:.2f}, C = {C_lattice:.2f}")
print(f"Small-World:  avg degree ≈ {avg_deg_small:.2f},   L = {L_small:.2f}, C = {C_small:.2f}")
```

## 3.4 Community Structure (Modularity)  

Networks often contain **modules or communities**. A common measure of modularity (Newman–Girvan) is:  

$$Q = \frac{1}{2m} \sum_{i,j} \left( A_{ij} - \frac{k_i k_j}{2m} \right) \delta(c_i, c_j),
$$ 

where $m$ is the total number of edges, $c_i$ is the community of node $i$, and $\delta$ is the Kronecker delta.  

**Example:** In citation networks, communities correspond to distinct research fields.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.community import greedy_modularity_communities
from networkx.algorithms.community.quality import modularity

# --- Reproducibility ---
seed = 42

# --- Graph A: Weak community structure (Erdős–Rényi) ---
N = 300
p_er = 0.03  # ~ average degree 9
G_er = nx.erdos_renyi_graph(n=N, p=p_er, seed=seed)

# Detect communities (greedy modularity)
comms_er = list(greedy_modularity_communities(G_er))
Q_er = modularity(G_er, comms_er)

# --- Graph B: Strong community structure (Stochastic Block Model) ---
sizes = [75, 75, 75, 75]  # 4 equal-sized groups
p_in = 0.12               # intra-community connection probability
p_out = 0.005             # inter-community connection probability
# Build block probability matrix
k = len(sizes)
P = [[p_out]*k for _ in range(k)]
for i in range(k):
    P[i][i] = p_in

G_sbm = nx.stochastic_block_model(sizes, P, seed=seed)

# Detect communities (greedy modularity)
comms_sbm = list(greedy_modularity_communities(G_sbm))
Q_sbm = modularity(G_sbm, comms_sbm)

# --- Helper: color nodes by community ---
def colors_by_communities(G, communities):
    # Assign an integer label to each detected community
    node_to_comm = {}
    for cid, comm in enumerate(communities):
        for v in comm:
            node_to_comm[v] = cid
    # Build a color array aligned with node order
    cmap = plt.cm.get_cmap('tab20')  # categorical palette
    colors = [cmap(node_to_comm[v] % cmap.N) for v in G.nodes()]
    return colors

colors_er = colors_by_communities(G_er, comms_er)
colors_sbm = colors_by_communities(G_sbm, comms_sbm)

# --- Layouts ---
pos_er = nx.spring_layout(G_er, seed=seed)
pos_sbm = nx.spring_layout(G_sbm, seed=seed)  # spring shows separated groups nicely

# --- Plot side-by-side ---
plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
nx.draw(
    G_er, pos_er,
    node_color=colors_er, node_size=22,
    edge_color="gray", width=0.3, alpha=0.9, with_labels=False
)
ax1.set_title(f"Weak Community Structure (ER)\nQ = {Q_er:.3f}, communities = {len(comms_er)}")
ax1.axis("off")

ax2 = plt.subplot(1, 2, 2)
nx.draw(
    G_sbm, pos_sbm,
    node_color=colors_sbm, node_size=22,
    edge_color="gray", width=0.3, alpha=0.9, with_labels=False
)
ax2.set_title(f"Strong Community Structure (SBM)\nQ = {Q_sbm:.3f}, communities = {len(comms_sbm)}")
ax2.axis("off")

plt.tight_layout()
plt.show()

deg_er = np.mean([d for _, d in G_er.degree()])
deg_sbm = np.mean([d for _, d in G_sbm.degree()])
print(f"ER   avg degree ≈ {deg_er:.2f}, modularity Q = {Q_er:.3f}, detected communities = {len(comms_er)}")
print(f"SBM  avg degree ≈ {deg_sbm:.2f}, modularity Q = {Q_sbm:.3f}, detected communities = {len(comms_sbm)}")
```

## 3.5 Assortativity (Degree Correlation)  

The **assortativity coefficient** $r$ measures whether nodes connect preferentially to nodes with similar degree:  

$$r = \frac{\sum_{jk} jk (e_{jk} - q_j q_k)}{\sigma_q^2},
$$ 

where $e_{jk}$ is the joint probability distribution of the degrees at the ends of an edge, $q_k$ is the normalized degree distribution, and $\sigma_q^2$ is its variance.  

**Example:** Social networks tend to be assortative, while technological networks are disassortative.  

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Reproducibility ---
seed = 42
rng = np.random.default_rng(seed)

# ---------------------------
# Assortative graph (by degree)
# ---------------------------
# Two dense communities -> nodes within each block have similar degrees
sizes = [150, 150]
p_in = 0.08     # within-block connection prob (dense)
p_out = 0.005   # cross-block prob (sparse)

k = len(sizes)
P = [[p_out]*k for _ in range(k)]
for i in range(k):
    P[i][i] = p_in

G_assort = nx.stochastic_block_model(sizes, P, seed=seed)

# Degree assortativity coefficient
r_assort = nx.degree_assortativity_coefficient(G_assort)

# Node coloring by degree (continuous colormap)
deg_assort = np.array([d for _, d in G_assort.degree()])
norm_assort = (deg_assort - deg_assort.min()) / (np.ptp(deg_assort) + 1e-12)
colors_assort = plt.cm.viridis(norm_assort)

# Layout
pos_assort = nx.spring_layout(G_assort, seed=seed)

# ------------------------------
# Disassortative graph (by degree)
# ------------------------------
# Random bipartite graph: high-degree nodes in one part tend to connect to
# lower-degree nodes in the other part -> negative assortativity
n_top, n_bottom = 120, 180
p_bip = 0.06

# Build bipartite sets
top_nodes = range(n_top)
bottom_nodes = range(n_top, n_top + n_bottom)

G_bip = nx.algorithms.bipartite.random_graph(n_top, n_bottom, p_bip, seed=seed)
G_dis = G_bip

r_dis = nx.degree_assortativity_coefficient(G_dis)

deg_dis = np.array([d for _, d in G_dis.degree()])
norm_dis = (deg_dis - deg_dis.min()) / (np.ptp(deg_dis) + 1e-12)
colors_dis = plt.cm.viridis(norm_dis)

# Layout: place bipartite layers apart for clarity
pos_dis = {}
x_gap = 1.8
for i, v in enumerate(top_nodes):
    pos_dis[v] = (-x_gap/2, i / max(1, n_top-1))  # left column
for j, v in enumerate(bottom_nodes):
    pos_dis[v] = ( x_gap/2, j / max(1, n_bottom-1))  # right column

# ------------------------------
# Plot side-by-side
# ------------------------------
plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
nx.draw(
    G_assort, pos_assort,
    node_color=colors_assort, node_size=24,
    edge_color="gray", width=0.3, alpha=0.9, with_labels=False
)
ax1.set_title(f"Assortative Mixing by Degree\nr = {r_assort:.3f}")
ax1.axis("off")

ax2 = plt.subplot(1, 2, 2)
nx.draw(
    G_dis, pos_dis,
    node_color=colors_dis, node_size=24,
    edge_color="gray", width=0.3, alpha=0.9, with_labels=False
)
ax2.set_title(f"Disassortative Mixing by Degree (Bipartite)\nr = {r_dis:.3f}")
ax2.axis("off")

plt.tight_layout()
plt.show()

print(f"Assortative graph:  avg degree ≈ {deg_assort.mean():.2f}, r = {r_assort:.3f}")
print(f"Disassortative graph: avg degree ≈ {deg_dis.mean():.2f}, r = {r_dis:.3f}")
```

## 3.6 Betweenness Centrality  

Measures how often a node lies on the **shortest paths** between other nodes:  

$$BC(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}},
$$  

where $\sigma_{st}$ is the number of shortest paths between $s$ and $t$, and $\sigma_{st}(v)$ is the number of those passing through $v$.  

**Example:** Airports like Dubai or Singapore often act as key transfer hubs.  


## 3.7 Closeness Centrality  

Measures how close a node is to all others in terms of path length:  

$$CC(v) = \frac{1}{\sum_{u \in V} d(v,u)}.
$$ 

**Example:** In communication networks, a node with high closeness centrality can spread information efficiently.  


## 3.8 Eigenvector Centrality  

Assigns importance to a node based on its connections to other important nodes:  

$$x_i = \frac{1}{\lambda} \sum_{j} A_{ij} x_j,
$$ 

where $\lambda$ is the largest eigenvalue of $A$.  

**Example:** Google’s **PageRank** is a variant of this measure.  


## 3.9 Network Density  

The **density** of a network quantifies how many edges are present compared to the maximum possible:  

$$D = \frac{2m}{N(N-1)},
$$  

where $m$ is the number of edges, and $N$ the number of nodes.  

**Example:** A complete graph has density $D=1$, while sparse real-world networks have low density.  


## 3.10 Diameter  

The **diameter** of a network is the maximum shortest path between any two nodes:  

$$D_{\text{max}} = \max_{i,j} d(i,j).
$$ 

**Example:** In a social network, this corresponds to the longest “chain” of acquaintances linking two people.  


## 3.11 Motifs  

A **network motif** is a small, recurring subgraph pattern that appears more frequently than expected in a random network.  

The **Z-score** is often used to measure motif significance:  

$$Z_m = \frac{N_m^{\text{real}} - \langle N_m^{\text{rand}} \rangle}{\sigma(N_m^{\text{rand}})},
$$

where $N_m^{\text{real}}$ is the number of motif $m$ in the real network, and the denominator normalizes against random expectation.  

**Example:** Feed-forward loops are common motifs in gene regulatory networks.

Beyond these classical metrics, network science also employs visualization techniques and computational algorithms to explore structure. Visualization allows researchers to see patterns—clusters, bridges, bottlenecks—that may not be obvious from raw data. Algorithms for community detection, motif finding, or network growth modeling provide more systematic ways of uncovering hidden organization.

Importantly, these tools are not only descriptive but predictive. By quantifying topological features, researchers can forecast how a network might respond to perturbations, such as the removal of a node, the introduction of new links, or the spread of a contagion. For example, analyzing degree distribution can reveal whether a system is robust against random failures but fragile against targeted attacks, while clustering coefficients can predict the likelihood of local information reinforcement in social systems.

In sum, the study of topology and the application of network metrics give us a lens to translate complexity into measurable form. Through this lens, we can detect regularities that cut across disciplines, compare systems that are otherwise incommensurable, and begin to understand the hidden order underlying seemingly chaotic interconnections.
